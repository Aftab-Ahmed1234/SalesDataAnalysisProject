```python
#Data
import pandas as pd

data = {
    "Product": ["Laptop", "Phone", "Laptop", "Tablet", "Phone", "Tablet"],
    "Month": ["Jan", "Jan", "Feb", "Feb", "Mar", "Mar"],
    "Sales": [50000, 30000, 60000, 25000, 40000, 35000],
    "Quantity": [5, 10, 6, 4, 12, 5]
}

df = pd.DataFrame(data)

df
```

```python
# Total Sales
print("Total Sales: ",df["Sales"].sum())
```

```python
# Average Sales
print("Average Sales: ",df["Sales"].mean())
```

```python
#Maximum Sales
print("Maximum Sales: ",df["Sales"].max())
```

```python
#Product-Wise Sales
print(df.groupby("Product")["Sales"].sum())
```

```python
#Mothly Sales
print(df.groupby("Month")["Sales"].sum())
```

```python
#Sales Bar Chart
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Product",y="Sales",data=df)
plt.xlabel("Product",color="red",size=15)
plt.ylabel("Sales",color="red",size=15)
plt.title("Product Sales", color="blue", size=20 )
plt.show()
```

```python
#Sales Line Chart
plt.plot(df["Month"], df["Sales"], marker="o")

plt.title("Monthly Sales")

plt.show()
```
